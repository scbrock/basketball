# Data Engineering

```python
# GET TEAM DATA
seasons = ['9697', '9798', '9899', '9900', '0001', '0102', '0203', '0304', '0405', '0506','0607', '0708', '0809', '0910', '1011', '1112', '1213', '1314', '1415', '1516', '1617', '1718']

# only change charlotte hornets before 0203
get_current_team = {'Vancouver Grizzlies': 'Memphis Grizzlies', 'New Jersey Nets': 'Brooklyn Nets', 'Charlotte Hornets': 'New Orleans Pelicans',
                    'New Orleans Hornets':'New Orleans Pelicans', 'Seattle SuperSonics':'Oklahoma City Thunder', 'Washington Bullets': 'Washington Wizards', 
                    'Charlotte Bobcats':'Charlotte Hornets', 'Charlotte Hornets 2': 'Charlotte Hornets'}

def build_team_data(seasons):
    root_name = 'teams/nba_team_'
    data = pd.DataFrame({})
    for season in seasons:
        file_name = root_name+season+'.csv'
        temp = pd.read_csv(file_name, index_col=0)
        temp['SEASON'] = [season]*temp.shape[0]
        data = pd.concat([data,temp], axis=0) # concatenate along rows

    return data

team_df = build_team_data(seasons)


# fix teams to current team names
def compare_helper(season):
    beginning = int(season[:2])
    year = None
    if beginning < 20:
        year = beginning + 2000
    else:
        year = beginning + 1900
    return year

def compare(season1, season2):
    if compare_helper(season1) < compare_helper(season2):
        return -1
    elif compare_helper(season1) == compare_helper(season2):
        return 0
    else:
        return 1

def old_to_current(data):
    teams = []
    for i, row in data.iterrows():
        team = row['TEAM']
        lkup = get_current_team.get(row['TEAM'])
        if team == 'Charlotte Hornets':
            if compare(row['SEASON'], '0203') == -1:
                teams.append(lkup)
            else:
                teams.append(team)
        else:
            if lkup:
                teams.append(lkup)
            else:
                teams.append(team)
    data['TEAM'] = teams
    return data
    
team_adj = old_to_current(team_df)
# print(sorted(team_adj['TEAM'].unique()))
# print(len(sorted(team_adj['TEAM'].unique())))

```

```python
# Get player data

def build_player_data(seasons):
    root_name = 'players/nba_player_'
    data = pd.DataFrame()
    for season in seasons:
        file_name = root_name+season+'.csv'
        temp = pd.read_csv(file_name, sep=',', index_col=0)
        temp = temp.dropna()
        teams = []
        for i, row in temp.iterrows():
            team = row['TEAM']
            teams.append(name_reverse_dictionary[team])
        temp['TEAM'] = teams
        temp['SEASON'] = [season]*len(teams)
        data = pd.concat([data, temp], axis=0)
        
    return data

player_df = build_player_data(seasons)
#print(player_df)
player_adj = old_to_current(player_df)
#print(len(player_adj['TEAM'].unique()))  # 30
```

# Feature Generation

```python
# to get target
def get_next_season(season):
    beginning = int(season[:2])
    year = None
    if beginning < 20:
        year = beginning + 2000
    else:
        year = beginning + 1900
    return str(year+1)[2:] + str(year+2)[2:]

# df[(df['col1'] >= 1) & (df['col1'] <=1 )]

#player_features = ['GP', 'PTS', 'REB', 'AST', 'TOV', 'STL', 'BLK', '+/-']
#player_features = ['PTS', 'AST', '+/-']
#team_features = ['W', 'PTS', '+/-', 'TOV', 'AST', 'PF']


def feature_gen_predict_player(target_stat='+/-'):
    feature_vectors = []
    target_vectors = []
    predict_features = player_features[:]
    #predict_features.remove(target_stat)
    #print(predict_features)
    ply_s_pairs = []
    
    for season in seasons[:-1]:
        isolate_season = player_adj[player_adj['SEASON']==season]
        for i, player in isolate_season.iterrows():
            target = player_adj[(player_adj['SEASON']==get_next_season(season)) & (player_adj['PLAYER']==player['PLAYER'])]
            if target.empty:
                continue
            feature_vector = list(player[player_features].values)
            target = list(target[player_features].values[0])
            feature_vectors.append(feature_vector)
            target_vectors.append(target)
            ply_s_pairs.append((player['PLAYER'], season))
            

    return feature_vectors, target_vectors, ply_s_pairs
    
X, y, pairs = feature_gen_predict_player()

# TO-DO: figure out saving and loading

print(time.time()-t1)
```

```python
# try to predict each feature
from xgboost import XGBRegressor
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression, Ridge, BayesianRidge

# Bayesian linear regression: https://towardsdatascience.com/introduction-to-bayesian-linear-regression-e66e60791ea7

# different model scoring: https://scikit-learn.org/stable/modules/model_evaluation.html
player_features = ['AGE', 'GP', 'W', 'L', 'MIN', 'PTS', 'FGM', 'FGA', 'FG%', '3PM', '3PA',
       '3P%', 'FTM', 'FTA', 'FT%', 'OREB', 'DREB', 'REB', 'AST', 'TOV', 'STL',
       'BLK', 'PF', 'FP', 'DD2', 'TD3', '+/-']

targets = player_features[:]

xgb_results = []
ridge_results = []
bayes_results = []
xgb_sd = []
ridge_sd = []
bayes_sd = []

for i, t in enumerate(targets):
    xgbr = XGBRegressor()
    ridge = Ridge()
    bayes = BayesianRidge()
    
    ts = [r[i] for r in y]
    kfold = KFold(n_splits=5, random_state=23)
    res_xgb = cross_val_score(estimator=xgbr, X=X, y=ts, scoring='neg_mean_squared_error' ,cv=kfold)
    res_ridge = cross_val_score(estimator=ridge, X=X, y=ts, scoring='neg_mean_squared_error' ,cv=kfold)
    res_bayes = cross_val_score(estimator=bayes, X=X, y=ts, scoring='neg_mean_squared_error' ,cv=kfold)
    
    xgb_results.append(res_xgb.mean())
    ridge_results.append(res_ridge.mean())
    bayes_results.append(res_bayes.mean())
    xgb_sd.append(res_xgb.std())
    ridge_sd.append(res_ridge.std())
    bayes_sd.append(res_bayes.std())
    
    #print(t,"XGB Explained Variance: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))

results_df = pd.DataFrame({'target': targets, 'xgb':xgb_results, 'xgb_sd':xgb_sd, 'ridge': ridge_results, 'ridge_sd':ridge_sd,'bayes': bayes_results, 'bayes_sd':bayes_sd})

#print(time.time()-t1)
```

```python
# Plot learning Curves

import numpy as np
import matplotlib.pyplot as plt
#from sklearn.naive_bayes import GaussianNB
#from sklearn.datasets import load_digits
from sklearn.model_selection import learning_curve
from sklearn.model_selection import ShuffleSplit
from xgboost import XGBRegressor


def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None,
                        n_jobs=None, train_sizes=np.linspace(.1, 1.0, 5)):
    """
    Generate a simple plot of the test and training learning curve.

    Parameters
    ----------
    estimator : object type that implements the "fit" and "predict" methods
        An object of that type which is cloned for each validation.

    title : string
        Title for the chart.

    X : array-like, shape (n_samples, n_features)
        Training vector, where n_samples is the number of samples and
        n_features is the number of features.

    y : array-like, shape (n_samples) or (n_samples, n_features), optional
        Target relative to X for classification or regression;
        None for unsupervised learning.

    ylim : tuple, shape (ymin, ymax), optional
        Defines minimum and maximum yvalues plotted.

    cv : int, cross-validation generator or an iterable, optional
        Determines the cross-validation splitting strategy.
        Possible inputs for cv are:
          - None, to use the default 3-fold cross-validation,
          - integer, to specify the number of folds.
          - :term:`CV splitter`,
          - An iterable yielding (train, test) splits as arrays of indices.

        For integer/None inputs, if ``y`` is binary or multiclass,
        :class:`StratifiedKFold` used. If the estimator is not a classifier
        or if ``y`` is neither binary nor multiclass, :class:`KFold` is used.

        Refer :ref:`User Guide <cross_validation>` for the various
        cross-validators that can be used here.

    n_jobs : int or None, optional (default=None)
        Number of jobs to run in parallel.
        ``None`` means 1 unless in a :obj:`joblib.parallel_backend` context.
        ``-1`` means using all processors. See :term:`Glossary <n_jobs>`
        for more details.

    train_sizes : array-like, shape (n_ticks,), dtype float or int
        Relative or absolute numbers of training examples that will be used to
        generate the learning curve. If the dtype is float, it is regarded as a
        fraction of the maximum size of the training set (that is determined
        by the selected validation method), i.e. it has to be within (0, 1].
        Otherwise it is interpreted as absolute sizes of the training sets.
        Note that for classification the number of samples usually have to
        be big enough to contain at least one sample from each class.
        (default: np.linspace(0.1, 1.0, 5))
    """
    plt.figure()
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("Score")

    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes, scoring='neg_mean_squared_error')
    print('train_scores:', train_scores)
    print('test_scores:', test_scores)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")

    plt.legend(loc="best")
    return plt


title = "Learning Curves XGBoost Regression Tree"
# Cross validation with 100 iterations to get smoother mean test and train
# score curves, each time with 20% data randomly selected as a validation set.
cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=23)

estimator = XGBRegressor(learning_rate= 0.05, max_depth= 3, random_state=23)
myplot = plot_learning_curve(estimator, title, X, target, ylim=(-100, 1), cv=cv, n_jobs=4)
myplot.show()

# title = r"Learning Curves (SVM, RBF kernel, $\gamma=0.001$)"
# # SVC is more expensive so we do a lower number of CV iterations:
# cv = ShuffleSplit(n_splits=10, test_size=0.2, random_state=0)
# estimator = SVC(gamma=0.001)
# plot_learning_curve(estimator, title, X, y, (0.7, 1.01), cv=cv, n_jobs=4)

# plt.show()
```
